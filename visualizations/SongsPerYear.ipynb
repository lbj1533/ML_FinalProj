# Year of Release Visualization

## Songs Per Year

Below is a visualization of the dataset which plots the number of songs per year. Given that this set includes the Top 100 songs from each year, it is expected to be 100 songs for each year.


```python
# importing libraries

import os
import pandas as pd
import glob

import matplotlib.pyplot as plt
```

```python
# loading data and showing first 3 rows

data_folder = "../data/raw/"
file = data_folder + "all_songs_data_processed.csv"

# concat all csvs into one dataframe
df_list = [pd.read_csv(file)]
df = pd.concat(df_list, ignore_index=True)
```

```python
# group by year and count songs
songs_per_year = df.groupby('Year').size().reset_index(name='song_count')

# sort by year
songs_per_year = songs_per_year.sort_values(by='Year')
```

```python
#plot num songs per year
plt.figure(figsize=(10, 6))
plt.plot(songs_per_year['Year'], songs_per_year['song_count'], marker='o', linestyle='-', color='b')
plt.title('Number of Songs per Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Songs')
plt.grid(True)
plt.show()
```

As you can see, the number is not consistent with the 100 a year intended value. However, the number does not dip below 90, and is thus very consistent. I do not anticipate this being an issue with our model, but it is good to realize it now.
